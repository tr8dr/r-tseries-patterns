# Get Data

```r
getSymbols("AAPL", from="2001-1-1", src='yahoo')

# reform into data frame
df = data.frame(time = index(AAPL), price = as.vector(AAPL[,'AAPL.Adjusted']))
```

```r
head(df)
```

# Evaluate Labeler

## minamp = 20%, Tinactive = 50 days

```r
obj <- AmplitudeBasedLabeler$new (minamp = 2000, Tinactive = 50)
labels <- obj$label (df[200:1000,], time = "time")
obj$plot(dimension = c(16,12))
```

## minamp = 20%, Tinactive = 25 days

```r
obj <- AmplitudeBasedLabeler$new (minamp = 2000, Tinactive = 25)
labels <- obj$label (df[200:1000,], time = "time")
obj$plot(dimension = c(16,12))
```
